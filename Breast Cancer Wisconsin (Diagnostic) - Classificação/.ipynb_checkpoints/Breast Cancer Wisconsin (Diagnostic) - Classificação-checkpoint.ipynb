```python
#importando pacotes necessários

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
sns.set_theme()
```

```python
#carregando dados
df = pd.read_csv('data.csv')
df.head()
```

```python
df.info()
```

A coluna Unnamed será dropada pois provavelmente ocorreu um erro durante o carregamento, pois ela não possui nenhum valor não nulo

```python
df.drop(columns = ['id','Unnamed: 32'], inplace = True)
```

Observando os dados do nosso dataset, podemos ver que contamos com uma só variável categórica, o diagnóstico clasifficado em benigno e maligno, e as outras variáveis são contínuas e trazem informações a cerca do tumor em análise.

Vamos inicar nosso projeto com uma EDA

# EDA

```python
#observar os valores das variáveis procurando se pode existir alguma anomalia nos valores
df.describe(percentiles = [.01,.25,.05,.75,.99])
```

Podemos observar que os calores máximos e mínimos estão longe dos percentis 99 e 1, respectivamente. Porém não podemos dizer categoricamente que se tratam de outliers neste momento

```python
#extrair os valores de y
y_vars = df.drop('diagnosis', axis=1).columns

fig, axes = plt.subplots(round(len(df.drop('diagnosis', axis=1).columns) / 3), 3, figsize=(12, 30))

for i, ax in enumerate(fig.axes):
    if i < len(df.drop('diagnosis', axis=1).columns):
        ax.set_xticklabels(ax.xaxis.get_majorticklabels())
        sns.boxplot(x='diagnosis', y=y_vars[i], data=df, ax=ax)
        ax.set_title('var ='+df.drop('diagnosis', axis=1).columns[i])

fig.tight_layout()
```

##  Insights

Tumores malignos tendem a apresentar:

1. Maior raio médio que os benignos
2. Maior perímetro médio e área
3. Possuem uma maior concavidade

Por outro lado, variáveis como: textura, suavidade, simetria_se, dimensão fractal não aparentam ser uma boa escolha para diferenciá-los pois os valores são muito próximos.

Próximo passo é analisar nossa variável target, ou seja, o diagnóstico



```python
#selecionando apenas a variável desejada
diagnosis_target = df['diagnosis']

#observar a distribuição dos valores
diagnosis_target.value_counts()
```

Temos em nosso dataset cerca de 63% de tumores benignos e 37% malignos. está desbalanceado, mas poderia ser pior.

```python
#utilizando um gráfico para melhor visualização da informação
fig,ax = plt.subplots()
ax = sns.countplot(diagnosis_target,alpha=0.7)
ax.set_title('Número de Tumores Malignos e Benignos')
ax.set_xlabel('Tipo do Tumor')
ax.set_ylabel('Quantidade')
plt.show()
```
