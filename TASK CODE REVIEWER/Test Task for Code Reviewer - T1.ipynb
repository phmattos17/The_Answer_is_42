    Q - The duration column contains data on the film length. How many missing values are there in this column?

```python
# change data type to numeric
#df_t1['duration'] = pd.to_numeric(df_t1['duration'], errors='coerce')
df_t1['duration'] = df_t1['duration'].astype('float64')
```

```python
duration_missing = df_t1['duration'].isna().sum()
print(f'There are {duration_missing} values on the duration column')
```

    Q - Replace the missing values in the duration column with the median value for this column

```python
# using the function fillna() alongside the np.median which calculate the median value for the column
#the usage of inplace = True, modify the orginal dataframe
df_t1['duration'].fillna(value = df_t1['duration'].median(),inplace = True)
```

    Q - What is the average film length? Give the answer as a floating-point figure rounded to two decimal places.

```python
mean_duration = df_t1['duration'].mean()
print(f'The average film lenght is {mean_duration:.2f} minutes')
```

    Q - Create a movie_duration_category column, which will contain three categories depending on the film length:
        â€¢ Category "1. <90" if the film is less than 90 minutes long
        â€¢ Category "2. 90â€“120" if the film is between 90 minutes and two hours long (inclusively)
        â€¢ Category "3. >120" if the film is more than two hours long

```python
#conditions to create the new column
conditions = [
    (df_t1['duration']<90),
    ((df_t1['duration']>=90) & (df_t1['duration']<=120)),
    (df_t1['duration']>120)
]

#values for each condition
category_name = [1,2,3]

df_t1['movie_duration_category'] = np.select(conditions,category_name)
```

    Q - Build a summary table for films released after 2000 (inclusively), to list the numbers of films:
        â€¢ Table rows: year
        â€¢ Table columns: movie duration category ("<90", "90â€“120", ">120")
        â€¢ The year of release should be displayed in the YYYY format.

```python
#filtering the orginal dataset
df_t1_after_2000 = df_t1[df_t1['title_year']>=2000]

#transform float to datatime format
df_t1_after_2000['title_year'] = pd.to_datetime(df_t1_after_2000['title_year'], format='%Y')

#create a column with only the realease year for each movie
df_t1_after_2000['year'] = pd.DatetimeIndex(df_t1_after_2000['title_year']).year

#group the columns by realease year and type
df_t1_after_2000.groupby(['year', 'movie_duration_category'])[
    'movie_duration_category'].count().to_frame(name='count')
```

    Q - The plot_keywords column holds keywords characterizing the film's plot. Using the data in this column, create a column called movie_plot_category, to contain four categories depending on the key words in the column:
    â€¢ Category "love_and_death" if the keywords include both "love" and "death"
    â€¢ Category "love" if the keywords include the word "love"
    â€¢ Category "death" if the keywords include the word "death"
    â€¢ Category "other" if the keywords do not meet the conditions above

```python
for index, row in df_t1.iterrows():
    try:
        keywords = row['plot_keywords'].split('|')
        if ('love' in keywords) & ('death' in keywords):
            df_t1.loc[index,'movie_plot_category'] = 'love_and_death'
        elif ('love' in keywords) & ('death' not in keywords):
            df_t1.loc[index,'movie_plot_category'] = 'love'
        elif (not 'love' in keywords) & ('death' in keywords):
            df_t1.loc[index,'movie_plot_category'] = 'death'
        else:
            df_t1.loc[index,'movie_plot_category'] = 'other'
    except:
        df_t1.loc[index,'movie_plot_category'] = 'other'
```

    Q - The imdb_score column shows a viewer rating for the film. Build a table to reflect the average rating of films depending on which movie_plot_category category they belong to.

```python
df_t1.groupby(['movie_plot_category'])['imdb_score'].mean().to_frame(name = 'avg_imdb_score')
```

    Q - What is the average rating of films in the "love" category? Give the answer as a floatingpoint figure rounded to two decimal places.

```python
mean_love_category = df_t1[df_t1['movie_plot_category'] == 'love']['imdb_score'].mean()
print(f'The average rating for the love category is {mean_love_category:.2f}.')
```
